```python
import matplotlib.pyplot as plt
arr=[1,2,3,4,5]
n=6
for i in range(len(arr)):
    if arr[i]==n:
       print("element found")
       break
else:
   print("element not found")
a=[1,2,3,4,5,6,7]
b=[10,20,30,40,6,60,70]
plt.plot(a,b)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("time complexity of linear search")
plt.show()

```
